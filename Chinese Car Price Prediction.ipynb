```python
#cp.cylindernumber.replace({'four':4,'six':6,'five':5,'three':3,'twelve':12,'two':2,'eight':8})
def num_map(x):
    if x== 'four':
        return 4  
    if x== 'six':
        return 6
    if x== 'five':
        return 5
    if x== 'three':
        return 3
    if x== 'tweleve':
        return 12   
    if x== 'two':
        return 2
    if x== 'eight':
        return 8
```

```python
cp['cylindernumber']=cp['cylindernumber'].apply(num_map)
```
